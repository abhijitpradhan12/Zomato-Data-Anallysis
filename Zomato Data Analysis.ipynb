### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
zomato=ds_zo.copy()
```

```python
# Prepair the rate column for analysis 
zomato['Rate'] = zomato['rate'].str.split('/').str[0]
zomato['Rate'].replace(['NEW','-'],['nan','nan'],inplace=True)
zomato['Rate'] = zomato['Rate'].astype(float)
```

```python
zomato['approx_cost(for two people)'].unique()
```

```python
# convert the "approx_cost(for two people)" column to float data type
zomato['approx_cost(for two people)'] = zomato['approx_cost(for two people)'].astype(str)
zomato['approx_cost(for two people)'] = zomato['approx_cost(for two people)'].apply(lambda x:x.replace(',',''))
zomato['approx_cost(for two people)'] = zomato['approx_cost(for two people)'].astype(float)
```

```python
zomato.dropna(subset=['Rate', 'approx_cost(for two people)'], inplace=True)
```

```python
zomato.info()
```

```python
zomato.drop(['rate','url','address','phone'],axis=1,inplace=True)
```

```python
zomato.head()
```

### Most popular Cuisines in Banglore

```python
def cuisine_cnt(zomato_df):
    cuisine_dic = {'Unknown':0}
    for j in range(len(zomato_df)):
        if zomato_df['cuisines'][j] is not np.nan:
            cus = zomato_df['cuisines'][j].split(', ')
            for new_cus in cus:
                if new_cus in cuisine_dic:
                    cuisine_dic[new_cus]+=1
                else:
                    cuisine_dic[new_cus]=1
        else:
            cuisine_dic['Unknown']+=1
    print("Total number of different cuisines types('unknown' included) :",len(cuisine_dic))
    print("-------------------------------------")
    return(cuisine_dic)
```

```python
cuisine_count = cuisine_cnt(zomato)
cuisine_count = pd.Series(cuisine_count)
print(cuisine_count.sort_values(ascending=False))
```

```python
fig11,ax =plt.subplots(figsize=(15,7))
sort_cuisines = cuisine_count.sort_values(ascending=False)
sb.barplot(sort_cuisines[:20].index,sort_cuisines[:20].values,ax=ax)
ax.set_title('Most popular Cuisines of Banglore',size=20)
ax.tick_params(axis='x',labelrotation=90)
ax.plot
```

### Top Cuisines in Each Location

```python
print('TOP CUISINE IN EACH LOCATION')

for loc in list(zomato['location'].unique()):
    print(loc)
    print('')
    loc_cui = cuisine_cnt(zomato[zomato['location']==loc].reset_index())
    loc_cui = pd.Series(loc_cui)
    print(loc_cui.sort_values(ascending=False).head())
    print("-------------------------------")
    
```
